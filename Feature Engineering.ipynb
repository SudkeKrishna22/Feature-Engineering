**Question 1:** What is a parameter?  

**Answer:**  
A parameter is a value that a model learns from the training data. For example, in linear regression, the slope and intercept are parameters.


**Question 2:** What is correlation?  
What does negative correlation mean?  

**Answer:**  
Correlation measures the strength and direction of a linear relationship between two variables. It ranges from -1 to 1.

Negative correlation means that as one variable increases, the other variable tends to decrease. For example, the more hours someone sleeps, the less tired they feel (negative correlation with fatigue).

**Question 4:** How does loss value help in determining whether the model is good or not?  

**Answer:**  
Loss measures the difference between predicted and actual values. A lower loss generally indicates a better model, but it must also be validated on unseen data to ensure generalization.


**Question 5:** What are continuous and categorical variables?  

**Answer:**  
- Continuous variables: Numeric variables that can take infinite values (e.g., height, weight).  
- Categorical variables: Variables that represent categories or labels (e.g., gender, color).


**Question 6:** How do we handle categorical variables in Machine Learning? What are the common techniques?  

**Answer:**  
- Label Encoding: Convert categories to numbers.  
- One-Hot Encoding: Create binary columns for each category.  
- Ordinal Encoding: Assign ordered integers if categories have ranking.

**Question 16:** What is an Optimizer?  
What are different types of optimizers? Explain each with an example.

**Answer:**  
An optimizer is an algorithm used to minimize the loss function by updating model parameters during training.
- **SGD (Stochastic Gradient Descent):** Updates weights using small batches.  
- **Adam:** Combines momentum and adaptive learning rates for faster convergence.  
- **RMSprop:** Adapts learning rates for each parameter, good for RNNs.

**Question 30:** Explain data encoding.  

**Answer:**  
Data encoding is the process of converting categorical variables into numerical form so that machine learning algorithms can process them. Common techniques include:  
- One-Hot Encoding: Creates binary columns for each category.  
- Label Encoding: Converts categories to integer labels.  
- Ordinal Encoding: Converts categories to ordered integers if there is a ranking.
